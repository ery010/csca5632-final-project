### Import libraries

```python
import os
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import itertools

from sklearn.decomposition import PCA
from sklearn.manifold import TSNE

from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture
from sklearn.metrics import silhouette_score

import warnings
warnings.filterwarnings("ignore")
```

### Load data

```python
# Load data
train = pd.read_csv("data/UNSW_NB15/UNSW_NB15_training-set.csv")
test = pd.read_csv("data/UNSW_NB15/UNSW_NB15_testing-set.csv")
```

The data set contains an `id` field and 2 labels, `attack_cat` and `label`. The `id` field is a unique identifier for each data point. The `attack_cat` field is a label that indicates the type of network attack and `label` specifies whether or not the traffic was benign or not.

```python
print(f"Train data size: {train.shape[0]} entries, {train.shape[1]-3} features")
train.head()
```

```python
print(f"Test data size: {test.shape[0]} entries, {test.shape[1]-3} features")
test.head()
```

```python
np.unique(train["attack_cat"])
```

The dataset contains 9 categories of network attacks and 1 category for normal traffic.
* `Normal`
* `Analysis`
* `Backdoor`
* `DoS`
* `Exploits`
* `Fuzzers`
* `Generic`
* `Reconnaissance`
* `Shellcode`
* `Worms`

Since we are building and training unsupervised models, we can utilize both the `train` and `test` sets for training.  

```python
# Merge data sets
full_data = pd.concat([train, test])
full_data.drop(columns=["id", "label"], inplace=True)
```

```python
full_data
```

### Data Cleaning

#### Check for duplicates

```python
# Check for duplicates
duplicates = full_data.duplicated()
print(f"Number of duplicates: {np.sum(duplicates)}")

# Drop duplicates
full_data = full_data.drop_duplicates()
```

#### Check for NaN values

```python
# Check for NaNs
nan_counts = full_data.isna().sum().sum()
print(f"Number of NaN values: {nan_counts}" )
```

Some of the features in the data are binary but have entries that have not been correctly inputted. Specifically, `is_ftp_login`, a feature that describes whether there was an FTP login attempt, should only be either a `0` or `1`. Entries that are not either will be treated as a binary 1 for our use case.

```python
full_data["is_ftp_login"].value_counts()
```

```python
def replace_binary(col):
    col = col.apply(lambda x: 1 if x >= 1 else 0)
    return col

full_data["is_ftp_login"] = replace_binary(full_data["is_ftp_login"])
```

```python
# Check that all values are 0 or 1
full_data["is_ftp_login"].value_counts()
```

### Exploratory Data Analysis

Let's look at the distribution of the attack categories.

```python
plt.figure(figsize=(12, 8))
sns.countplot(y=full_data["attack_cat"], data=full_data["attack_cat"])
plt.xlabel("Count")
plt.ylabel("Attack category")
plt.title("Distribution of attack categories")
```

We can drop the `attack_cat` column since it will not be used for building the models. We will store it separately for later use.

```python
data_labels = full_data["attack_cat"]
full_data.drop(columns=["attack_cat"], inplace=True)
```

We will need to convert string features into binary representations using one-hot encoding. One-hot encoding generates additional features (columns) for each possible value, encoding `1` if the feature matches the value, otherwise `0`.

```python
onehot_features = full_data.columns[full_data.dtypes == 'object']
onehot_features
```

```python
full_data = pd.get_dummies(full_data , columns=onehot_features)
full_data
```

### Dimensionality Reduction using PCA
We can use the `PCA()` class to reduce the number of components in the data. To determine the number of components, we can analyze the relationship between the number of PCA components and variance. We want to select a number of components that preserves as much variance as possible.

```python
# Function for displaying variance against number of components
def plot_pca_var(data):
    pca = PCA().fit(data)
    plt.figure(figsize=(8, 8))
    plt.plot(np.cumsum(pca.explained_variance_ratio_))
    plt.xlabel("Number of PCA Components")
    plt.ylabel("Variance")
    plt.title("Variance vs. Number of Components")
    plt.grid()
    plt.show()
```

```python
plot_pca_var(full_data)
```

We can see that most of the variance is captured by very small number of PCA components. Note that this does **not** mean that only a small number of the original features are considered in the model. It just means that the information from the original features could be compressed into a small number of PCA components.

We can set `n_components` as the variance we want captured instead of a direct value.

```python
pca = PCA(n_components=0.95)
pca_data = pca.fit_transform(full_data)
```

### Model development and training

#### K-Means Clustering

```python
kmeans = KMeans(n_clusters=10, init="random")
kmeans_labels = kmeans.fit_predict(pca_data)
```

Now let's look at the clustering results from the `KMeans` model. We can do this in several ways.

#### Evaluate performance with Silhouette Score

```python
silhouette_score(pca_data, kmeans_labels)
```

The silhouette_score of `0.55511` suggests that the clusters are mostly well-separated. This means that there are some cluster overlaps but are distinct in general. We can visualize the clustering by projecting the PCA data on a 2-D scatterplot.

#### Projection into a 2-D scatterplot

```python
tsne = TSNE(n_components=2, perplexity=30, random_state=42)
data_2d = tsne.fit_transform(pca_data)
```

```python
plt.figure(figsize=(8, 8))
sns.scatterplot(x=data_2d[:, 0], y=data_2d[:, 1],
                palette="viridis",
                hue=kmeans_labels)
plt.show()
```

Now let's try a different model. We will train a `GaussianMixture` model on the same data.

```python
gmm = GaussianMixture(n_components=10, covariance_type="full")
gmm_labels = gmm.fit_predict(pca_data)
```

```python
silhouette_score(pca_data, gmm_labels)
```

The `GaussianMixture` model performed similarly (slightly worse) to the `KMeans` model. Let's project the PCA data using the `GaussianMixture` labels.

```python
plt.figure(figsize=(8, 8))
sns.scatterplot(x=data_2d[:, 0], y=data_2d[:, 1],
                palette="viridis",
                hue=gmm_labels)
plt.show()
```

The scatterplots are very similar, with some slight differences in cluster assignments towards the center of the data points. This is consistent with the silhouette scores.

### Fine-tuning KMeans
The shape of the scatterplot has the approximate shape of a sphere, so `KMeans` is likely more appropriate than `GaussianMixture` for this problem. Let's try to fine-tune some of the parameters to try to improve the silhouette score.

```python
clusters = [3, 5, 7, 10]
init_var = ["random", "k-means++"]

best_silhoutte_score = 0
best_n_clusters = 0
best_init = None

for n_clusters in clusters:
    for init in init_var:
        kmeans = KMeans(n_clusters=n_clusters, random_state=42, init=init)
        score = silhouette_score(pca_data, kmeans_labels)
        if score > best_silhoutte_score:
            best_silhoutte_score = score
            best_n_clusters = n_clusters
            best_init = init

print(f"Best KMeans parameters: n_clusters={n_clusters}, init={init}")
print(f"Silhouette score: {best_silhoutte_score}")
```

The best model for `KMeans` was the model with the parameters `init="k-means++` and `n_clusters=10`. It performed slightly better than the initial model. `KMeans` requires the number of clusters to be preset, and one of the motivations for tuning the model with less clusters was to investigate for potential similarities between the different types of attacks (in which case the model may perform better with less clusters).

### Results and Summary

|       Model          | Silhouette Score |
| -------------------- |----------------- |
|        `KMeans`      |     0.55511      |
| `KMeans (optimized)` |     0.55827      |
|   `GaussianMixture`  |     0.51423      |

The results suggest that clustering methods like `KMeans` could be useful for differentiating network attacks. In the cybersecurity space, it is important to understand different types of attacks because they do not behave the same and we cannot rely on "one-size-fits-all" solutions. The possible advantage of deploying unsupervised models like `KMeans` and `GaussianMixture` over supervised methods is that unsupervised models can potentially learn patterns that arise from new types of attacks. Specifically, unsupervised models like clustering are not restricted by labels that enforce a prediction, meaning they can possibly detect attacks that do not fit a standard set of characteristics. We can also learn overlaps between different types of attacks, which can help determine more optimal allocation of resources for prevention and response strategies. Deploying unsupervised models is also more efficient and scalable in real-time, since they require less processing steps like labeling.
